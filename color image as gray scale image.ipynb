# Load color image as gray scale image 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("dog-image.jpg")
plt.imshow(img)
```

```python
# read image as a gray scale
img_new = cv2.imread("dog-image.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img_new)
```

```python
plt.imshow(img_new, cmap= "gray")
```
