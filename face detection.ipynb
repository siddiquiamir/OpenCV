```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('butterfly.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img1= cv2.imread('beatriz.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

group= cv2.imread('group1.jpg')
group= cv2.cvtColor(group, cv2.COLOR_BGR2RGB)
```

```python
# this has around 6 features of face and we are going to pass to this to our image to see if those features are in our image
# and whether it is able to detect the face or not
face_cascade = cv2.CascadeClassifier('C:/Users/amirs/Anaconda3/pkgs/libopencv-4.7.0-py311h787696b_6/Library/etc/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detection(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y), (x+w, y+h), (0,255,0),10)
    return face_img
```

```python
result1 = face_detection(img)
plt.imshow(result1)
```

```python
result2 = face_detection(img1)
plt.imshow(result2)
```

```python
result3 = face_detection(group)
plt.imshow(result3)
```

```python
def advanced_face_detection(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y), (x+w, y+h), (0,255,0),10)
    return face_img
```

```python
result2 = advanced_face_detection(img1)
plt.imshow(result2)
```

```python
result3 = advanced_face_detection(group)
plt.imshow(result3)
```
