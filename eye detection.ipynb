```python
# this has around 6 features of face and we are going to pass to this to our image to see if those features are in our image
# and whether it is able to detect the face or not
eye_cascade = cv2.CascadeClassifier('C:/Users/amirs/Anaconda3/pkgs/libopencv-4.7.0-py311h787696b_6/Library/etc/haarcascades/haarcascade_eye.xml')
```

```python
def eye_detection(img):
    face_img = img.copy()
    eye_rects = eye_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img,(x,y), (x+w, y+h), (0,255,0),10)
    return face_img
```
