# Resizing image

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("dog-image.jpg")
type(img)
```

```python
plt.imshow(img)
```

```python
img_new = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_new)
```

```python
img_new.shape
```

```python
img_res = cv2.resize(img_new, (1500, 1000))
plt.imshow(img_res)
```

```python
img_res.shape
```

```python
img_new.shape
```

```python
w_ratio = 0.3
h_ratio = 0.3

img_1 = cv2.resize(img_new, (0,0), img_new, w_ratio, h_ratio)
plt.imshow(img_1)
```

```python
# Flip image
img_flip = cv2.flip(img_new, -1)
plt.imshow(img_flip)
```
